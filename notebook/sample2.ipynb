```python
# load tocsv module
from yard.utils.tocsv import dataframes_from_db
```

```python
# get data from database
start_date = '2014-08-10'
end_date = '2014-08-13'
column = 'dt, bid_price1, ask_price1'

codes = ['KORBIT', 'BITSTAMP', 'BTCCHINA', 'USDKRW', 'CNYKRW',]
sum_df = dataframes_from_db(codes, column, start_date, end_date)
bitstamp_bid_krw_base = sum_df.BITSTAMP_bid_price1 * sum_df.USDKRW_bid_price1
bitstamp_ask_krw_base = sum_df.BITSTAMP_ask_price1 * sum_df.USDKRW_bid_price1
btcchina_bid_krw_base = sum_df.BTCCHINA_bid_price1 * sum_df.CNYKRW_bid_price1
btcchina_ask_krw_base = sum_df.BTCCHINA_ask_price1 * sum_df.CNYKRW_bid_price1

# type(btcchina_ask_krw_base)
# print btcchina_ask_krw_base

sum_df['bitstamp_bid_krw_base'] = bitstamp_bid_krw_base
sum_df['bitstamp_ask_krw_base'] = bitstamp_ask_krw_base
sum_df['btcchina_bid_krw_base'] = btcchina_bid_krw_base
sum_df['btcchina_ask_krw_base'] = btcchina_ask_krw_base

extract_df = sum_df[['KORBIT_bid_price1', 'bitstamp_bid_krw_base', 'btcchina_bid_krw_base']]
# print sum_df.columns
print extract_df

# print sum_df
```
